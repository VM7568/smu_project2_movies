```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import requests
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.imdb.com/chart/top?sort=ir,desc&mode=simple&page=1'
browser.visit(url)
```

```python
html = browser.html
imdb = BeautifulSoup(html)
type(imdb)
```

```python
top_movies = imdb.find("table", {"data-caller-name":"chart-top250movie"}).find("tbody").find_all("tr")
len(top_movies)
```

```python
top_movies
```

```python
test = top_movies[0]
title = test.find("td", {"class": "titleColumn"}).find("a").text
title
```

```python
link = test.find("td", {"class": "titleColumn"}).find("a")["href"]
link
```

```python
rating = test.find("td", {"class": "imdbRating"}).find("strong").text
rating
```

```python
ratingText = test.find("td", {"class": "imdbRating"}).find("strong")["title"]
ratingText
```
